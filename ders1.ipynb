```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
data = pd.read_csv("data_with_nans.csv")
data.head()
```

```python
data.info()
```

```python
data.drop(labels=[data.columns[0]],axis=1,inplace=True)
```

```python
data.describe()
```

```python
data.groupby("Species").agg(["min","max","std","mean"])
```

```python
data.isna().sum()
```

```python
for columns in data.columns[1:-1]:
    data[columns].fillna(value=data[columns].mean(),inplace=True)
```

```python
data.isna().sum()
```

```python
data.dropna(axis=0,how="any",inplace=True)
```

```python
sbn.scatterplot(data=data,x="Id",y="SepalLengthCm",hue="Species")
```

```python
for i in data.columns[1:-1]:
    sbn.scatterplot(data=data,x="Id",y=i,hue="Species")
    plt.show()
```

```python
for columns in data.columns[1:-1]:
    for spec in data["Species"].unique():
        selected_spec=data[data["Species"]==spec]
        selected_column=selected_spec[columns]

        std=selected_column.std()
        avg=selected_column.mean()

        tree_sigma_plus=avg+(std*3)
        tree_sigma_minus=avg-(std*3)

        outliers=selected_column[((selected_spec[columns]>tree_sigma_plus)|(selected_spec[columns]<tree_sigma_minus))].index
        data.drop(index=outliers,inplace=True)
        print(outliers)
```

```python
for i in data.columns[1:-1]:
    sbn.scatterplot(data=data,x="Id",y=i,hue="Species")
    plt.show()
```

```python
for column in data.columns[1:-1]:
    for spec in data["Species"].unique():
        selected_spec=data[data["Species"]==spec]
        selected_column=selected_spec[column]

        q1=selected_column.quantile(0.25)
        q3=selected_column.quantile(0.75)

        iqr=q3-q1

        min=q1-(1.5*iqr)
        max=q3+(1.5*iqr)

        min_idxs=data[(data["Species"]==spec)&(data[column]<min)].index
        print(min_idxs)
        max_idxs=data[(data["Species"]==spec)&(data[column]>max)].index
        print(max_idxs)

        data.drop(index=min_idxs,inplace=True)
        data.drop(index=max_idxs,inplace=True)
```

```python
for i in data.columns[1:-1]:
    sbn.scatterplot(data=data,x="Id",y=i,hue="Species")
    plt.show()
```

```python
data.to_csv("final_data.csv")
```
